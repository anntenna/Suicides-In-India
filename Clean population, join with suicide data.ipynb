```python
import pandas as pd
from pandas import DataFrame, Series
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
pop1 = DataFrame.from_csv('population_by_education_gender_2001.csv')
pop1.head()
```

```python
pop1 = pop1[pop1.Gender.isin(['Female','Male'])]
```

```python
pop1.index = pop1.index.str.replace('Orissa','Odisha')
```

```python
pop2 = DataFrame.from_csv('population_by_education_gender_2011.csv')
pop2.head()
```

```python
pop1.Type = pop1.Type.str.title()
pop1.Type.unique()
```

```python
pop2.Type = pop2.Type.str.title()
pop2.Type.unique()
```

```python
pop2.Type = pop2.Type.str.replace('Literate Without Education Level','Literate Without Educational Level')
pop2.Type = pop2.Type.str.replace('Higher Secondary/Intermediate Pre-University/Senior Secondary','Higher Secondary/Intermediate/Pre-University/Senior Secondary')
pop2.Type = pop2.Type.str.replace('Illiterates','Illiterate')
pop2.Type = pop2.Type.str.replace('Literates','Literate')
pop2.Type = pop2.Type.str.replace('Population Attending Educational Institution','Population Attending Educational Institutions')
pop1.Type = pop1.Type.str.title()
pop2.Type = pop2.Type.str.title()
```

```python
pop2.head()
```

```python
pop1.reset_index(inplace=True)
pop1.set_index(['State','Gender','Type'],inplace=True)
pop2.reset_index(inplace=True)
pop2.set_index(['State','Gender','Type'],inplace=True)
```

```python
pop1.rename(index=str, columns={'Total':'2001'},inplace=True)
pop2.rename(index=str, columns={'Total':'2011'},inplace=True)
```

```python
pop1.drop('Year',axis=1,inplace=True)
pop2.drop('Year',axis=1,inplace=True)
```

```python
pop = pd.concat([pop1,pop2],axis=1)
```

```python
pop.head()
```

```python
pop['diff'] = pop['2011'] - pop['2001'] 
```

```python
import math
pop['diff'] = pop['diff'].apply(lambda x: math.ceil(x/10))
```

```python
pop['2014'] = pop['2013'] + pop['diff']
```

```python
pop.drop(['diff'],axis=1,inplace=True)
```

```python
pop = DataFrame(pop.stack()).reset_index().rename(index=str,columns={'level_3':'Year',0:'Total'})
```

```python
pop.Gender.unique()
```

```python
pop.to_csv('population_stats_by_education.csv')
```
