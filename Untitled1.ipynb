```python
pop = DataFrame.from_csv('education_distribution_states_2011.csv')
pop.head()
```

```python
pop['Area Name'].unique()
```

```python
pop = pop[~(pop['Area Name'] == 'INDIA')]
pop['Area Name'] = pop['Area Name'].str.replace('State - ','').str.replace(r'  \d.*','').str.title()
```

```python
pop['Age-group'].unique()
```

```python
pop = pop[pop['Total/ Rural/ Urban'] == 'Total']
```

```python
pop = pop[pop['Age-group'] == 'All ages']
```

```python
pop.drop(['State Code','Distt. Code','Total/ Rural/ Urban','Age-group'],axis=1,inplace=True)
```

```python
pop.head()
```

```python
pop = pop.reset_index().drop(['Table Name'],axis=1)
```

```python
pop.rename(index=str, columns={'Area Name':'State'},inplace=True)
```

```python
# df.columns = pd.MultiIndex.from_product([df.columns, ['C']])

pop = pop.set_index(['State'])

pop_male = pop[pop.columns[pop.columns.str.contains('Males')]]
pop_male.columns = pd.MultiIndex.from_product([['Male'],pop_male.columns])

pop_female = pop[pop.columns[pop.columns.str.contains('Females')]]
pop_female.columns = pd.MultiIndex.from_product([['Female'],pop_female.columns])

pop_total = pop[pop.columns[pop.columns.str.contains('Persons')]]
pop_total.columns = pd.MultiIndex.from_product([['Total'],pop_total.columns])
```

```python
pop = pd.concat([pop_male,pop_female,pop_total], axis=1)
pop.head()
```

```python
pop = DataFrame(pop.stack(level=0).stack()).reset_index()
pop.head()
```

```python
pop.rename(index=str, columns = {'level_1':'Gender','level_2':'Type',0:'Total'}, inplace=True)
pop.head()
```

```python
pop['Type'] = pop.Type.str.replace(' - Persons','')
pop['Type'] = pop.Type.str.replace('Educational level - ','')
pop['Type'] = pop.Type.str.replace(' - Females','')
pop['Type'] = pop.Type.str.replace(' - Males','')
```

```python
pop.head()
```

```python
pop.State.unique()
```

```python
pop.State = pop.State.str.replace('Andaman & Nicobar Islands','A & N Islands')
pop.State = pop.State.str.replace('Nct Of Delhi','Delhi (Ut)')
pop.State = pop.State.str.replace('Dadra & Nagar Haveli','D & N Haveli')
```
